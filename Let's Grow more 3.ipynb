Exploratory Data Analysis on a terrorism Dataset

```python
#importing packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the terrorism dataset
dataset_path = 'terrorism.csv'
data = pd.read_csv(dataset_path, encoding='latin1')  # Adjust encoding if needed

# Display basic information about the dataset
print(data.info())
print(data.head())

# Summary statistics
print(data.describe())

# Plot a bar chart of the number of incidents per year
plt.figure(figsize=(12, 6))
sns.countplot(x='iyear', data=data)
plt.xticks(rotation=45)
plt.title('Number of Terrorist Incidents Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.show()

# Plot a pie chart of attack types
plt.figure(figsize=(8, 8))
data['attacktype1_txt'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Attack Types')
plt.ylabel('')
plt.show()

# Create a heatmap of correlations between numeric columns
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Plot a box plot of fatalities by attack type
plt.figure(figsize=(12, 6))
sns.boxplot(x='attacktype1_txt', y='nkill', data=data)
plt.xticks(rotation=45)
plt.title('Fatalities by Attack Type')
plt.xlabel('Attack Type')
plt.ylabel('Number of Fatalities')
plt.show()
```
